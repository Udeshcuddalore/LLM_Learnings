Sentence Transformers on Hugging Face

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "this is a test document"
query_results = embeddings.embed_query(text)
query_results
```

```python
len(query_results)
```

```python
doc_result = embeddings.embed_documents([text, "This is not a test document."])
doc_result[1]
```

using hugging face embeddings -- sentence_transformers

```python

from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Define your sentences
sentences = [
    "The cat sits outside", 
    "A man is playing guitar", 
    "The feline is resting outdoors"
]

# 3. Encode the sentences into vectors
embeddings = model.encode(sentences)
```

```python
score = util.cos_sim(embeddings[0], embeddings[2])
print(f"Similarity Score: {score.item():.4f}")
```
