# Character Text splitter

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size = 100, chunk_overlap = 20)
text_splitter.split_documents(docs)
```

```python
speech  = " "
with open("speech.txt") as f:
    speech = f.read()
text_splitter = CharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
text = text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```

# HTML Header Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string = """
<body>
    <header>
        <h1>The Great HTML Splitter Test</h1>
        <p>Testing <strong>metadata preservation</strong> and hierarchy.</p>
    </header>

    <main>
        <section id="introduction">
            <h2>1. Structural Depth</h2>
            <div>
                <p>This is a standard paragraph. It contains enough text to act as a primary chunk for most splitting logic.</p>
                <div>
                    <p>This paragraph is <em>nested</em> inside an extra div to test recursive depth handling.</p>
                </div>
            </div>
        </section>

        <hr />

        <section id="formatting">
            <h2>2. Inline Elements and Lists</h2>
            <p>We want to ensure that <code>&lt;code&gt; tags</code> and <a href="#">hyperlinks</a> don't get sliced in half mid-tag. Here is a list to check block-level splitting:</p>
            <ul>
                <li>First item with <span style="color: blue;">inline styling</span>.</li>
                <li>Second item with a significantly longer string of text to see if the splitter forces a break within a list element or keeps the LI intact.</li>
                <li>Third item.</li>
            </ul>
        </section>

        <section id="tables">
            <h2>3. Tabular Data</h2>
            <table border="1">
                <thead>
                    <tr>
                        <th>Feature</th>
                        <th>Status</th>
                    </tr>
                </thead>
                <tbody>
                    <tr>
                        <td>Context Awareness</td>
                        <td>High Priority</td>
                    </tr>
                    <tr>
                        <td>Tag Integrity</td>
                        <td>Required</td>
                    </tr>
                </tbody>
            </table>
        </section>

        <section id="edge-cases">
            <h2>4. Empty Tags and Comments</h2>
            <p>The following div is empty:</p>
            <div class="empty-container"></div>
            <p>Final sentence to mark the end of the main content area.</p>
        </section>
    </main>

    <footer>
        <p>&copy; 2026 Splitter Lab. All rights reserved.</p>
    </footer>
</body>
"""
```

```python
headers_to_split_on = [("h1","Header 1"), ("h2","Header 2")]
```

```python
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
```

```python
import requests
url = "https://plato.stanford.edu/entries/goedel/"
html_content = requests.get(url).text
headers_to_split_on = [("h1","Header 1"), ("h2","Header 2")]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
chunks = html_splitter.split_text(html_content)
chunks
```

# Recursive JSON Splitter

```python
import json
import requests
json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
splitter = RecursiveJsonSplitter(max_chunk_size=600)
```

```python
json_chunks = splitter.split_json(json_data=json_data)
for chunk in json_chunks[:3]:
    print(type(chunk))
```

```python
docs = splitter.create_documents(texts = [json_data])
for doc in docs[:3]:
    print(doc)
```

```python
texts = splitter.split_text(json_data=json_data)
print(texts[0])
print(texts[300])
```
