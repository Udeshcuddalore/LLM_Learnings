- Build a simple application with LangChain

```python
import os
from dotenv import load_dotenv
load_dotenv()

```

```python
#import chatopenAI
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o") #Model Names: As of 2026, you generally want to use gpt-4o (Omni) or gpt-4-turbo, as they are faster and more capable than the original gpt-4.
print(llm)
```

```python
## Input and get response from LLM 
result = llm.invoke("What is generative AI?")
```

```python
print(result)
```

```python
### chatprompt template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "you are an expert AI Engineer, Provide me answers based on the questions"),
        ("human", "{input}")
    ]
)
prompt
```

```python
# 3. Create the Chain (The Pipe!)
chain = prompt | llm
```

```python
reponse = chain.invoke({"input": "can you tell me about hyderabad city"})
print(reponse)
```

```python
## strouput parser

from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt|llm|output_parser
reponse = chain.invoke({"input": "can you tell me about hyderabad city"})
print(reponse)
```
